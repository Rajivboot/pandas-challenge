## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Binning & grouping purchase_data data frame by age

bins=[0, 9, 14, 19, 24, 29, 35, 40, 50]     
group_labels=["<10", "10-14", "15-19", "20-24","25-29","30-34","35-39","40+"]

#slice and place the data in new_row

df["Age Ranges"] = pd.cut(df["Age"], bins, labels=group_labels)

# remove duplicates
user1_df = df.drop_duplicates(["SN"]) 

#Total count
new_group = user1_df.groupby("Age Ranges")["SN"].count()


#Percentage of Players
new_group1 = new_group.sum()
new_groupper = new_group/new_group1
pd.options.display.float_format = '{:,.2%}'.format

#data Framing
Age_Demographics=pd.DataFrame(
    {"Total Count": new_group,"Percentage of Players": new_groupper,}
     )
Age_Demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Purchase Count 
Player = df["SN"].value_counts()
#Average Purchase Price
Player1=df.groupby("SN")["Price"].mean()
#Total purchse
Player2=df.groupby("SN")["Price"].sum()

#Framing
topspenders=pd.DataFrame(
    {"Purchase_Count": Player, "Average Purchase Price": Player1,"Total Purchase Value": Player2})

#Sorting
topspenders = topspenders.sort_values("Total Purchase Value", ascending=False)

#Rounding
topspenders["Total Purchase Value"] = topspenders["Total Purchase Value"].map("${:,.2f}".format)
topspenders["Average Purchase Price"] = topspenders["Average Purchase Price"].map("${:,.2f}".format)

topspenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create data frame with items related information 
items = df[["Item ID", "Item Name", "Price"]]

# Group the item double data 
item_stats = items.groupby(["Item ID","Item Name"])

# Count the number of times an item has been purchased 
purchase_count_item = item_stats["Price"].count()

# Calcualte the purchase value per item 
purchase_value = (item_stats["Price"].sum()) 

# Find individual item price
item_price = purchase_value/purchase_count_item

# Create data frame with obtained values
most = pd.DataFrame({"Purchase Count": purchase_count_item, 
                                   "Item Price": item_price,
                                   "Total Purchase Value":purchase_value})

# Sort in descending order to obtain top spender names and provide top 5 item names
most = most.sort_values(["Purchase Count"], ascending=False)

# Format with currency style
most.style.format({"Item Price":"${:,.2f}", "Total Purchase Value":"${:,.2f}"})
                                                    
most
```

* Sort the above table by total purchase value in descending order
